```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from support_vector import HardMarginSVC
from sklearn.svm import LinearSVC
```

```python
data = pd.read_csv("Data.csv")
data
```

```python
x = data.drop(["class"], axis=1)
y = data["class"]
```

```python
y = np.where(y==0, -1, 1)
y
```

```python
scaler = StandardScaler()
x_std = scaler.fit_transform(x)
```

```python
model = HardMarginSVC()
model.fit(x_std, y)
```

```python
plt.scatter(x_std[:,0],x_std[:,1],c = y)
xx1, xx2 = np.meshgrid(
    np.arange(-2,2,.01),
    np.arange(-2,2,.01)
)
meshgred_points = np.array([xx1.ravel(),xx2.ravel()]).T
z = model.predict(meshgred_points)
z = z.reshape(xx1.shape)
plt.contourf(xx1, xx2, z,alpha = .2)
```

```python
model1 = LinearSVC()
model1.fit(x_std, y)
```

```python
plt.scatter(x_std[:,0],x_std[:,1],c = y)
xx1, xx2 = np.meshgrid(
    np.arange(-2,2,.01),
    np.arange(-2,2,.01)
)
meshgred_points = np.array([xx1.ravel(),xx2.ravel()]).T
z = model1.predict(meshgred_points)
z = z.reshape(xx1.shape)
plt.contourf(xx1, xx2, z,alpha = .2)
```
